## Import

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing import image_dataset_from_directory
# from keras import Sequential
# from keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense
import seaborn as sns
import matplotlib.pyplot as plt
# from random import randint
# from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

## Data Training

```python
train_set = image_dataset_from_directory(
    "dataset/train",
    labels="inferred",
    label_mode="binary",
    class_names=["Negative", "Positive"],
    image_size=(64, 64),
    shuffle=True,
    seed=1,
    subset="training",
    validation_split=0.1,
    batch_size=128
)
```

```python
n_train = 0
n_neg_train = 0
n_pos_train = 0

for image, label in train_set.unbatch():
    n_train += 1
    
    if (label==0):
        n_neg_train += 1
    else:
        n_pos_train += 1
        
print(n_train, n_neg_train, n_pos_train)
```

```python
x1 = "Negative"
y1 = n_neg_train

x2 = "Positive"
y2 = n_pos_train

plt.figure(figsize =(7, 7))
plt.bar(x1, y1, color='red')
plt.bar(x2, y2, color='blue')
plt.xlabel("data Positif dan Negatif")
plt.ylabel("Keseluruhan data train")
plt.title("Banyaknya data Training")
plt.show()
```

## Data Validasi

```python
val_set = image_dataset_from_directory(
    "dataset/train",
    labels="inferred",
    label_mode="binary",
    class_names=["Negative", "Positive"],
    image_size=(64, 64),
    shuffle=True,
    seed=1,
    subset="validation",
    validation_split=0.1,
    batch_size=128
)
```

```python
n_val = 0
n_neg_val = 0
n_pos_val = 0

for image, label in val_set.unbatch():
    n_val += 1
    
    if (label==0):
        n_neg_val += 1
    else:
        n_pos_val += 1
        
print(n_val, n_neg_val, n_pos_val)
```

```python
x1 = "Negative"
y1 = n_neg_val

x2 = "Positive"
y2 = n_pos_val

plt.figure(figsize =(7, 7))
plt.bar(x1, y1, color='red')
plt.bar(x2, y2, color='blue')
plt.xlabel("data Positif dan Negatif")
plt.ylabel("Keseluruhan data validasi")
plt.title("Banyaknya data Validasi")
plt.show()
```
